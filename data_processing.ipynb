```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import json, os
import subprocess
```

```python
# json for data make requests
my_json = {
    "name": None,
    "x_0": None,
    "v_0": None,
    "omega": None,
    "dt": None,
    "end_time": None,
    "parameters": {
        "jsonpath": None,
        "datapath": None
    }
}
```

```python
# make_data function makes json file there main data store. 
# This json file is given to binary file with "binaryfile_path" path. 
# Binary file must make datafile path to which must be saved in json file.
# make_data function returns path to json file

def make_data() -> str:
    # changeable part
    exp = my_json.copy()
    exp["name"] = input("experiment name =")
    exp["x_0"] = float(input("x_0 ="))
    exp["v_0"] = float(input("v_0 ="))
    exp["omega"] = float(input("omega ="))
    exp["dt"] = float(input("dt ="))
    exp["end_time"] = float(input("time of the end ="))

    # unchangeable part
    jsonpath = "jsonfiles/"+ exp["name"] + ".exp" # here json will be
    binaryfile_path = "data_maker.out" # binary file of datamaker

    # changebeable part
    exp["parameters"]["jsonpath"] = jsonpath

    # unchangeable part
    with open(jsonpath, "w") as f:
        json.dump(exp, f, indent=2)
    p = subprocess.Popen([f"./{binaryfile_path}", f"{jsonpath}"])
    a = p.communicate()
    return jsonpath
```

```python
def experiment(x):
    data = pd.read_csv(f"datafiles/{x}.txt", sep="\t")
    t = data.iloc[:, 0]
    x = data.iloc[:, 1]
    t = t.to_numpy()
    x = x.to_numpy()

    plt.figure(figsize=(16, 9))
    
    X = np.arange(0, 20, 0.01)
    plt.plot(X, np.sin(X))
    
    plt.xlabel('$t$', fontsize=14)
    plt.ylabel('$x$', fontsize=14)
        
    plt.scatter(t, x, marker="s", color="black", s=1)
    plt.grid()
    plt.show()    
```
